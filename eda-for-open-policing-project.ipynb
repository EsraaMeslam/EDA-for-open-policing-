<h1>Visualizations

```python
df.hist(figsize=(20,20),color='blue') # Visualizations distribution columns that have a data type "Number "
plt.show()
```

```python
gender_counts = df['driver_gender'].value_counts()
gender_counts.plot(kind='bar', color=['blue', 'hotpink'], edgecolor='black')

plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
(df[df['violation']=='Speeding']['driver_gender'].value_counts()).plot(kind='bar',color='blue',edgecolor='black')
plt.title(' Men Vs Women stopped')
plt.show()
```

```python
sns.kdeplot(df['driver_age'], color='blue',fill=True)

plt.title('Driver Age Distribution (Density Plot)')
plt.xlabel('Driver Age')
plt.ylabel('Density')
plt.show()
```

```python
print ('The most Freq Value in "driver_age" col "Mode" is : ',df['driver_age'].mode() ) #22 is the most freq val
```

```python

sns.kdeplot(df['driver_age_raw'], color='blue',fill=True)

plt.title('Driver Age Raw Distribution (Density Plot)')
plt.xlabel('Driver Age Raw')
plt.ylabel('Density')
plt.show()
```

```python
print ('The most Freq Value in "driver_age_raw" col "Mode" is : ',df['driver_age_raw'].mode() )
```
